## Transform

```python
infla = pd.DataFrame.from_dict(infla.items())
```

```python
infla.replace('...', '', inplace=True)
infla
```

```python
infla.drop(index=0, inplace=True)
```

```python
infla = infla.rename(columns={0: 'AnoMes', 1:'I'})
infla['Dia'] = '15'
infla = infla.reset_index(drop=True)
```

```python
infla['DATA'] = infla['AnoMes'].str.cat(infla['Dia'])
```

```python
infla = infla[['DATA','I']]
```

```python
infla['DATA'] = pd.to_datetime(infla['DATA'])
```

```python
infla['INDICE'] = 0
infla.head()
```

```python
last_item = int(len(infla)-1)
print(last_item)
infla['INDICE'][last_item] = 1
```

```python
infla1 = infla.sort_values('DATA', ascending=False, ignore_index=True)
```

```python
infla1.info()
```

```python
infla1['I'] = infla1['I'].astype(float)
for i in range(1, len(infla1)):
  infla1.loc[i , 'INDICE'] = infla1.loc[i - 1 , 'INDICE'] / (1 + infla1.loc[i - 1, 'I'] / 100)
```

```python
infla1['D'] = 0
infla1.head()
```

```python
infla1['D'][0] = 1
for i in range(1, len(infla1)):
  infla1.loc[i , 'D'] = infla1.loc[0 , 'INDICE'] / infla1.loc[i , 'INDICE']
```
